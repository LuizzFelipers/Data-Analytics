Neesa análise, vou utilizar dados do IBGE. Foi um levantamento feito para avaliar os setores empregados no país entre os anos de 2015 até 2025.

Os setores são:

- Público;

- Privado no formato CLT e sem carteira assinada;

- Autônomos;

- Pessoas no mercado de trabalho.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df_geral = pd.read_csv("dados.csv", sep=";", encoding="Latin-1")

df_geral
```

```python
df_geral = df_geral.rename(columns={
    '24370 - Pessoas em idade de trabalhar - PNADC - Unidades (mil)' : 'Pessoas em Idade de trabalhar (mil)',
    '24371 - Empregados no setor privado e público - PNADC - Unidades (mil)' : 'Empregados no setor público e privado (mil)',
    '24372 - Empregados no setor público - PNADC - Unidades (mil)' : 'Empregados no setor público (mil)',
    '24373 - Empregado por conta própria - PNADC - Unidades (mil)' : 'Autonomos (mil)',
    '24375 - Empregados no setor privado - Com carteira - PNADC - Unidades (mil)' : 'Empregados no Setor Privado com carteira assinada (mil)',
    '24376 - Empregados no setor privado - Sem Carteira - PNADC - Unidades (mil)' : 'Empregados no Setor Privado sem carteira assinada (mil)'
})

df_geral
```

```python
df_geral = df_geral.drop(124)

df_geral
```

```python
converter_numero = ['Pessoas em Idade de trabalhar (mil)',
                    'Empregados no setor público e privado (mil)',
                    'Empregados no setor público (mil)',
                    'Autonomos (mil)',
                    'Empregados no Setor Privado com carteira assinada (mil)',
                    'Empregados no Setor Privado sem carteira assinada (mil)']


df_geral[converter_numero] = (
    df_geral[converter_numero]
    .apply(pd.to_numeric, errors='coerce')
    .astype('float')  # 'Int64' permite valores nulos
)

df_geral.info()
```

```python
df_geral["Total de empregados no Setor Privado com e sem carteira assinada (mil)"] = df_geral['Empregados no Setor Privado com carteira assinada (mil)'] + df_geral['Empregados no Setor Privado sem carteira assinada (mil)']

df_geral
```

```python
df_geral["Data"] = pd.to_datetime(df_geral['Data'],format='%m/%Y')
```

```python
df_geral['Ano'] = df_geral['Data'].dt.year
```

```python
df_anual = df_geral.groupby('Ano').mean(numeric_only=True).reset_index()
```

```python
df_geral = df_geral.drop(columns={"Data"})

df_geral
```

```python
df_geral.info()
```

```python
df_anual = df_geral.groupby('Ano').mean().reset_index()

df_anual = round(df_anual,2)

df_anual
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo do gráfico
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Plotar todas as colunas (exceto 'Ano')
colunas = df_anual.columns.drop('Ano')
for coluna in colunas:
    sns.lineplot(data=df_anual, x='Ano', y=coluna, label=coluna, marker='o')

# Ajustes do gráfico
plt.title('Comparativo Anual de Indicadores Trabalhistas', fontsize=14)
plt.xlabel('Ano')
plt.ylabel('Valor Médio')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Legenda fora do gráfico
plt.xticks(df_anual['Ano'].unique())  # Garantir que todos os anos apareçam
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
#amplitude setor privado
amplitude_em_idade_trabalho = df_anual["Total de empregados no Setor Privado com e sem carteira assinada (mil)"].iloc[10] - df_anual["Total de empregados no Setor Privado com e sem carteira assinada (mil)"].iloc[0]
amplitude_em_idade_trabalho = round(amplitude_em_idade_trabalho,2)

#amplitude setor público e privado
amplitude_setor_publico_privado = df_anual["Empregados no setor público e privado (mil)"].iloc[10] - df_anual["Empregados no setor público e privado (mil)"].iloc[0]
amplitude_setor_publico_privado = round(amplitude_setor_publico_privado,2)

#Amplitude setor público 
amplitude_setor_publico = df_anual["Empregados no setor público (mil)"].iloc[10] - df_anual["Empregados no setor público (mil)"].iloc[0]
amplitude_setor_publico = round(amplitude_setor_publico,2)

#Amplitude Autônomos 
amplitude_autonomos = df_anual["Autonomos (mil)"].iloc[10] - df_anual["Autonomos (mil)"].iloc[0]
amplitude_autonomos = round(amplitude_autonomos,2)

#Amplitude do setor privado 
amplitude_CLT_privado = df_anual["Total de empregados no Setor Privado com e sem carteira assinada (mil)"].iloc[10] - df_anual['Total de empregados no Setor Privado com e sem carteira assinada (mil)'].iloc[0]
amplitude_CLT_privado = round(amplitude_CLT_privado,2)

amplitudes_geral = {"Variação de pessoas(em 10 anos) em idade para trabalhar(mil)":[amplitude_em_idade_trabalho],
                    "Variação de pessoas(10 anos) do setor público e privado(mil)":[amplitude_setor_publico_privado],
                    "Variação de pessoas(10 anos) do setor público(mil)":[amplitude_setor_publico],
                    "Variação de pessoas(10 anos) de autônomos(mil)":[amplitude_autonomos],
                    "Variação de pessoas(10 anos) do setor privado com ou sem carteira assinada":[amplitude_CLT_privado]
                    }

df_amplitudes = pd.DataFrame(amplitudes_geral)

df_amplitudes
```

```python
df_melted = df_amplitudes.T.reset_index()
df_melted.columns = ["Categoria", "Variação (mil)"]  

plt.style.use("dark_background")

plt.figure(figsize=(20, 15))
sns.barplot(
    data=df_melted,
    x="Categoria",
    y="Variação (mil)",
    palette="viridis"
)


plt.title("Variação de Pessoas em 10 Anos (em milhares)", fontsize=28)
plt.xlabel("Categoria",fontsize = 25)
plt.ylabel("Variação (mil)",fontsize = 25)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

Insights:


O setor Público foi o que menos empregou pessoas durante o período de 10 anos.

Em contrapartida, o setor privado liderou o número de pessoas empregadas

ALém disso, o PNADC registrou um aumento significativo de, aproximadamente, 6 milhões. Isso significa que a expectativa de vida do Brasil aumentou.

Uma boa justificativa para o setor privado estar á frente do público pode ser a alta demanda por empregados, uma vez que o Brasil ultrapassou, segundo o IBGE, os 64 milhões de CNPJ's cadastrados. Sendo que o setor de microempresas foi o líder.

Já o número de autônomos pode ser explicado pelas novas modalidades das empresas de tecnologia contratarem prestadores de serviços.
Um bom exemplo seria as empresas IFOOD, Uber, 99-Pop, Mercado Livre e outras que contratam entregadores como autônomos com pouca barreira de entrada.

Outrossim, o Brasil, de forma geral, conseguiu empregar mais trabalhadores do que formou profissionais. E isso é um ponto interessante e mostra uma boa evolução.

Entretanto, o sub-emprego, empregos com alta carga horária, salários baixos e péssimas condições de trabalho, devem ser superados.
